**IMPORT LIBRARY**

```python
import matplotlib.pyplot as plt
```

**DATASET**

```python
#DATA SAMPLE
months = ('Jan', 'Feb', 'Mar', 'Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec')
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20,22, 30, 25, 35, 40, 45]
product_C_sales = [18, 22, 25, 28, 32, 38, 42, 45, 48,52, 56, 60]
product_D_sales = [7, 9, 11, 13, 16, 18,20, 22, 24, 26, 28, 30]
```

```python
#Line Plot
plt.plot(months, product_A_sales, label='Produk A')
plt.plot(months, product_B_sales, label='Produk B')
plt.title('Penjualan Produk A Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.show()
```

**LINE PLOT**

```python
plt.plot(months, product_A_sales, label='Produk A')
plt.plot(months, product_B_sales, label='Produk B')
plt.title('Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.show()
```

**KUSTOMISASI**

```python
plt.plot(months, product_A_sales, label='Produk A', color='blue', linestyle='--', marker='o')
plt.plot(months, product_B_sales, label='Produk B', color='red', linestyle='-', marker='x')
plt.title('Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**MULTIPLE LINES**

```python
plt.plot(months, product_A_sales, label='Produk A', marker='o')
plt.plot(months, product_B_sales, label='Produk B', marker='x')
plt.title('Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.show()
```

**GARIS BERBEDA UNTUK MENUNJUKKAN TREND**

```python
plt.plot(months, product_A_sales, label='Produk A Trend', linestyle='--', color='blue')
plt.plot(months, product_B_sales, label='Produk B Trend', linestyle='-', color='red')
plt.plot(months, product_C_sales, label='Produk C Trend', linestyle='dashed', color='purple')
plt.plot(months, product_D_sales, label='Produk D Trend', linestyle='solid', color='green')
plt.title('Trend Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Subplot untuk Menampilkan Beberapa Grafik**

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

#Product A
axs[0].plot(months, product_A_sales, label='Product A', color="blue", marker='o')
axs[0].set_title('Product A Sales')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

#Product B
axs[1].plot(months, product_B_sales, label='Product B', color="red", marker='x')
axs[1].set_title('Product B Sales')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
